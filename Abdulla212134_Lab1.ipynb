```python
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
```


# 1.) Import a colored image from your computer (Lab1_1.jpg).

```python
colored_image=cv.imread('/content/Lab1_1.jpg')

#convert image from HSV to RGB
rgb_image = cv.cvtColor(colored_image, cv.COLOR_BGR2RGB)

#display image
plt.imshow(rgb_image)
```

# 2.) Find the size of your image.

```python
height, width, channel = colored_image.shape

print("Width: ", width)
print("Height: ", height)
print("Channel: ", channel)
```

# 3.) Resize the image (Lab1_1.jpg) with three different resolutions (100 x 100 x 3), (200 x 200 x 3), (400 x 400 x 3). Discuss the impact of resizing on image quality and computational complexity.

```python
#resizing image to (100x100x3)
resized_image_1=cv.resize(rgb_image,(100,100))

plt.imshow(resized_image_1)
```

```python
#resizing image to (200x200x3)
resized_image_2=cv.resize(rgb_image,(200,200))

plt.imshow(resized_image_2)
```

```python
#resizing image to (400x400x3)
resized_image_3=cv.resize(rgb_image,(400,400))

plt.imshow(resized_image_3)
```

```python
''' In terms of quality, resizing the image will affect the resolution and aspect ratio of the image.
    The original image has a 1920x1080 which a 16:9 aspect ratio. Resizing the image resulted in
    reducing the pixel density (downscaling) and changed the aspect ratio to 1:1, making it look more
    compressed.


    In terms of computational complexity, reducing the image quality speeds up the processing due to
    fewer pixels. This is useful in simple computer vision applications, but in applications that
    require accurate data, downscaling is not an advantage.'''
```

# 4.) Create a colored image with resolution (320 x 320 x 3) and set all pixels to be black then save it with name (Black.jpg).

```python
# data type unit8 (unsigned 8-bit integer) meaning each pixel has an intensity range from 0 to 255
black_image = np.zeros((320, 320, 3), dtype="uint8")
plt.imshow(black_image)
cv.imwrite('Black.jpg', black_image)
```

# 5.) Create the same picture as (2) but set all pixels to be white and save it as (White.jpg).

```python
white_image = np.ones((320, 320, 3), dtype="uint8")*255
plt.imshow(white_image)
cv.imwrite('White.jpg', white_image)
```

# 7.) Do the following operation to 𝑓(𝑢,𝑣) and explain the effects of each operation.

**a.)** $2𝑓(𝑢,𝑣)$

```python
task7_image=cv.imread('/content/abdulla212134.jpg')
task7_image_rgb = cv.cvtColor(task7_image, cv.COLOR_BGR2RGB)

product1 = cv.multiply(task7_image_rgb, 2)
plt.imshow(product1)
#The contrast is reduced in the image, since white blends in with the blue.
#The value of the blue channel is 120 so multipling it by two makes it 240,
#which is close to 255 (max value).
#As a result, there is less contrast between the blue text and the white background.
```

**b.)** $\frac{𝑓(𝑢,𝑣)}{2}$







```python
product2 = cv.divide(task7_image_rgb, 2)
plt.imshow(product2)
#the brightness of the image decreases
```

**c.)** $𝑓(𝑢,𝑣) +100$

```python
array3 = cv.add(task7_image_rgb,100)
plt.imshow(array3)
#the brightness of the image increases
#the black edges become gray because it when from 0 to 100
```

**d.)** $𝑓(𝑢,𝑣) -100$

```python
array4 = cv.subtract(task7_image_rgb,100)
plt.imshow(array4)
#the brightness of the image decreases
```

**e.)** $255−𝑓(𝑢,𝑣)$

```python
array5 = cv.subtract(255,task7_image_rgb)
plt.imshow(array5)
#image is inverted
#The white pixels (255) become zero and the black pixels (0) become 255
```


**f.)** $0.3f_R + 0.6f_G + 0.1f_B$


```python
channel_R = task7_image_rgb[:,:,0]
channel_G = task7_image_rgb[:,:,1]
channel_B = task7_image_rgb[:,:,2]

# channel_R1 = cv.multiply((channel_R),0.3)
# channel_G1 = cv.multiply((channel_G),0.6)
# channel_B1 = cv.multiply((channel_B),0.1)
# channel_RGB = cv.add((channel_R1),(channel_G1),(channel_B1))
channel_RGB = (0.3*channel_R + 0.6*channel_G + 0.1*channel_B)
plt.imshow(channel_RGB, cmap='gray')

#This formula is convert a color image to grayscale
#because it combines the three channels (RGB) into a single intensity value.
```
